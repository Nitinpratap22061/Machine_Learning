# Naive Bayers for classification problem

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data
```

```python
data = data[data['species']!='setosa']
data
```

```python
species= pd.get_dummies(data['species'],drop_first=True)
```

```python
data.drop(['species'],axis=1,inplace=True)
data
```

```python
dataset = pd.concat([data,species],axis=1)
dataset
```

```python
dataset['virginica'] = dataset['virginica'].map({False:0,True:1})
dataset
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
pred = gnb.fit(X_train,y_train).predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_test,pred)
score
```
