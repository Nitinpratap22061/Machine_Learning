```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
# creating value
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2D')
```

```python
y = x*x
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2D')
```

```python
pi = np.pi
x = np.arange(0,4*pi,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2D')
```

```python
x = np.arange(0,4*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y_sin,'r--')
plt.title('sine graph')
plt.subplot(1,2,2)
plt.plot(x,y_cos,'b--')
plt.title('cosine graph')
```

## Bar diagram

```python
x = [1,8,20]
y = [2,14,3]
x1 = [2,24,15]
y1 = [3,6,12]

plt.bar(x,y,color='r')
plt.bar(x1,y1)
```

## pie chart


```python
label = ['bihar','up','bengal','rajasthan']
sizes = [231, 432, 124, 250 ]
color = ['gold','green','red','yellow']
explode = [0.1,0,0,0]
plt.pie(sizes,explode=explode,labels=label,colors=color ,autopct='%1.1f%%',shadow=True)
```
