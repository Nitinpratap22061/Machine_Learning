```python
total_score_df=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
#but we want first inning total
```

```python
total_score_df=total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df.head()
```

```python
match_df=match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df['team1'].unique()
```

```python
current_teams=[
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Chennai Super Kings',
    'Delhi Capitals',
    'Rajasthan Royals',
    'Kings XI Punjab', 
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')


```

```python
match_df = match_df[match_df['team1'].isin(current_teams)]
match_df = match_df[match_df['team2'].isin(current_teams)]
```

```python
match_df.head()
```

```python
match_df = match_df[match_df['dl_applied']==0]
```

```python
match_df.head()
```

```python
match_df['dl_applied'].value_counts()
```

```python
match_df
```

```python
delivery.head()
```

```python
# chahiye kya match_df se uth lo
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
match_df
```

```python
delivery_df=match_df.merge(delivery,on='match_id') 
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df
```

```python
## think ki run_left kaise nikale hr ball ke baad
```

```python

delivery_df['curr_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
#Now for runs left:-
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['curr_score']
```

```python
delivery_df
```

```python
## balls left ke liye thou12
```

```python
delivery_df['balls_left']=126-(delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
## wickets left:-
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else"1")
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
wickets
```

```python
delivery_df['wickets_left'] = 10-wickets
```

```python
delivery_df.head(40)
```

```python
# current run rate wala bnao
```

```python
delivery_df['crr'] = delivery_df['curr_score'] * 6 / (120-delivery_df['balls_left'])
```

```python
delivery_df
```

```python
# rrr
```

```python
delivery_df['rrr'] = delivery_df['runs_left']*6/delivery_df['balls_left']
```

```python
delivery_df.head()
```

```python
# winner ke liye 0 1
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
    
    
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1) #axis =1 mtlb row
```

```python
delivery_df
```

```python
final_df=delivery_df[['batting_team' , 'bowling_team' , 'city','runs_left','balls_left','total_runs_x','wickets_left','crr','rrr','result']]
final_df.dropna(inplace=True)
final_df=final_df[final_df['balls_left'] !=0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state =1 )
```

```python
X_train
```

```python
# ohe on categorical data
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf' , OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
trf
```

```python
final_df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1' , trf),
    ('step2' ,LogisticRegression(solver='liblinear'))
])
```

```python
final_df
```

```python
pipe.fit(X_train,y_train)
```

```python
X_train.describe()
```

```python
X_train
```

```python
trf
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```
