# SEABORN
## DISTRIBUTION PLOT
- distplot
- joinplot => bivariate analysis
- pairplot  => more than two features analyis

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

# correlation with heatmap
- to find relation between two features

```python
df['total_bill'].corr(df['tip'])
```

```python
sns.jointplot(data=df, x='total_bill', y='tip',kind='reg')
```

```python
sns.pairplot(df,hue='sex')
```

```python
df
```

```python
sex = pd.get_dummies(df['sex'],drop_first=True)
sex
```

```python
smoker = pd.get_dummies(df['smoker'],drop_first=True)
smoker
```

```python
day = pd.get_dummies(df['day'],drop_first=True)
day
```

```python
time = pd.get_dummies(df['time'],drop_first=True)
time
```

```python
df.drop(['sex','day','smoker','time'],axis=1,inplace=True)
df
```

```python
df = pd.concat([df,sex,day,smoker,time],axis=1)
df
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```
